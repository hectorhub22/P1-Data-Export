Generación del documento .csv

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Semilla para reproducibilidad
np.random.seed(42)
random.seed(42)

# Definir listas de posibles valores
productos = ['Manzanas', 'Vino Cabernet Sauvignon', 'Salmón Congelado', 'Aceite de Oliva', 'Arándanos Frescos', 'Muebles de Madera', 'Alpaca Textil', 'Cobre Refinado', 'Truchas Frescas', 'Jugo de Frutas', 'Nueces', 'Espárragos Frescos', 'Quinoa', 'Carne de Vacuno', 'Flores Frescas', 'Café Arábica', 'Camarones Congelados', 'Textiles de Algodón', 'Plátanos', 'Queso Gouda']
unidades_medida = ['Kilogramos', 'Botellas', 'Litros', 'Unidades', 'Metros', 'Toneladas', 'Docenas', 'Sacos']
paises_destino = ['Estados Unidos', 'Reino Unido', 'Japón', 'Italia', 'Países Bajos', 'México', 'Australia', 'China', 'Canadá', 'Brasil', 'Alemania', 'España', 'Francia', 'Corea del Sur', 'Colombia', 'Argentina', 'Perú', 'Bélgica', 'Suiza', 'Suecia']
puertos_origen_chile = ['Valparaíso', 'San Antonio', 'Puerto Montt', 'Talcahuano', 'Arica', 'Antofagasta', 'Iquique', 'Punta Arenas']
tipos_transporte = ['Marítimo', 'Aéreo', 'Terrestre']

# Función para generar una fecha aleatoria dentro de un rango
def random_date(start_date, end_date):
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + timedelta(days=random_number_of_days)
    return random_date.strftime('%Y-%m-%d')

# Generar los datos para 1500 registros
data = []
for i in range(1, 1501):
    fecha = random_date(datetime(2023, 1, 1), datetime(2025, 4, 29))
    producto = random.choice(productos)
    cantidad = np.random.randint(100, 10000)
    unidad = random.choice(unidades_medida)
    pais = random.choice(paises_destino)
    ciudad = f"Ciudad_{pais.replace(' ', '')}_{np.random.randint(1, 5)}" # Ciudades ficticias
    puerto_origen = random.choice(puertos_origen_chile)
    puerto_destino = f"Puerto_{pais.replace(' ', '')}_{np.random.randint(1, 5)}" # Puertos ficticios
    transporte = random.choice(tipos_transporte)
    distancia = np.random.randint(500, 25000)
    costo_envio = distancia * np.random.uniform(0.05, 0.3) # Costo proporcional a la distancia
    valor_producto = cantidad * np.random.uniform(1, 50)
    impuestos = valor_producto * np.random.uniform(0.01, 0.05)
    cliente = f"Cliente_{np.random.randint(1, 100)}" # Clientes ficticios
    direccion = f"Calle {np.random.randint(100, 999)}, Barrio {np.random.randint(1, 20)}, Ciudad {pais.replace(' ', '')}" # Direcciones ficticias

    data.append([f"EXP{i:04d}", fecha, producto, cantidad, unidad, pais, ciudad, puerto_origen, puerto_destino, transporte, distancia, costo_envio, valor_producto, impuestos, cliente, direccion])

# Crear el DataFrame de pandas
df_exportaciones = pd.DataFrame(data, columns=['ID_Exportacion', 'Fecha_Exportacion', 'Producto', 'Cantidad', 'Unidad_Medida', 'Pais_Destino', 'Ciudad_Destino', 'Puerto_Origen', 'Puerto_Destino', 'Tipo_Transporte', 'Distancia_KM', 'Costo_Envio_USD', 'Valor_Producto_USD', 'Impuestos_Exportacion_USD', 'Cliente', 'Direccion_Cliente'])

# Guardar el DataFrame en un archivo CSV
df_exportaciones.to_csv('exportaciones_sinteticas_1500.csv', index=False)

print("Se ha generado el archivo 'exportaciones_sinteticas_1500.csv' con 1500 registros.")
```
